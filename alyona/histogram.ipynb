```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def count_unique_pairs_in_directory(directory_path):
    node_pair_counts = []
    
    # Loop through CSV files in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith(".csv"):
            movie_csv_path = os.path.join(directory_path, filename)

            try:
                df_movie_nodes = pd.read_csv(movie_csv_path)
                
                # Drop duplicate rows based on 'node_1' and 'node_2' columns to get unique pairs
                unique_pairs = df_movie_nodes[['node_1', 'node_2']].drop_duplicates()
                num_unique_pairs = len(unique_pairs)
                node_pair_counts.append(num_unique_pairs)
            except FileNotFoundError:
                print(f"CSV file {filename} not found!")
            except KeyError:
                print(f"CSV file {filename} does not contain the required 'node_1' or 'node_2' columns!")
    
    return node_pair_counts

def plot_histogram(node_pair_counts, mean_value, variance_value, std_deviation_value):
    #CHANGE
    # Compute the frequency of each unique node pair count
    unique_counts, frequencies = np.unique(node_pair_counts, return_counts=True)
    
    plt.figure(figsize=(10, 6))
    plt.bar(unique_counts, frequencies, color='skyblue', edgecolor='black')
    plt.xlabel('Number of Unique Node Pairs')
    plt.ylabel('Number of Movies')
    plt.title('Unique Node Pairs Distribution in Bollywood Least Popular')

    # Add the metrics as text on the plot
    metrics_text = f"Mean: {mean_value:.2f}\nVariance: {variance_value:.2f}\nStd Dev: {std_deviation_value:.2f}"
    plt.text(0.95, 0.95, metrics_text, fontsize=12, verticalalignment='top', horizontalalignment='right', 
             transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))
    
    plt.show()

def calculate_statistics(node_pair_counts):
    mean_value = np.mean(node_pair_counts)
    variance_value = np.var(node_pair_counts)
    std_deviation_value = np.std(node_pair_counts)
    
    return mean_value, variance_value, std_deviation_value

def main():
    directory_path = input("Enter the path to the directory containing movie CSV files: ")
    
    if not os.path.isdir(directory_path):
        print("Directory not found! Please provide a valid directory path.")
        return
    
    node_pair_counts = count_unique_pairs_in_directory(directory_path)
    
    if node_pair_counts:
        mean_value, variance_value, std_deviation_value = calculate_statistics(node_pair_counts)
        plot_histogram(node_pair_counts, mean_value, variance_value, std_deviation_value)

if __name__ == "__main__":
    main()
```
