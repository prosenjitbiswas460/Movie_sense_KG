```python
import os
import pandas as pd
import shutil

base_dir = '/mnt/Data/prabirmondal/prabir/python_program/movie_sense/SRI_KG/Movie_sense_KG/Movie_sense_KG/knowledge_graph'
new_dir = '/mnt/Data/prabirmondal/prabir/python_program/movie_sense/SRI_KG/Movie_sense_KG/Movie_sense_KG/alyona'

def create_new_dir_structure():
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
    for category in ['bollywood', 'hollywood']:
        for subcategory in ['Least popular', 'popular']:
            path = os.path.join(new_dir, category, subcategory)
            os.makedirs(path, exist_ok=True)

def process_files():
    for category in ['bollywood', 'hollywood']:
        for subcategory in ['Least popular', 'popular']:
            sub_dir = os.path.join(base_dir, category, subcategory)
            for file in os.listdir(sub_dir):
                if file.endswith('.csv'):
                    file_path = os.path.join(sub_dir, file)
                    df = pd.read_csv(file_path)

                    # Check if any edge has 3 or more words
                    df['word_count'] = df['edge'].apply(lambda x: len(str(x).split()))
                    edge_word_count = df[df['word_count'] >= 3]
                    
                    # If any row satisfies the condition, move the file
                    if not edge_word_count.empty:
                        relative_path = os.path.relpath(file_path, base_dir)
                        new_file_path = os.path.join(new_dir, relative_path)

                        # Ensure the directory structure exists
                        new_file_dir = os.path.dirname(new_file_path)
                        os.makedirs(new_file_dir, exist_ok=True)

                        # Copy the file to the new directory
                        shutil.copy(file_path, new_file_path)
                        print(f"Copied: {file_path} to {new_file_path}")

if __name__ == "__main__":
    create_new_dir_structure()
    process_files()
```
